```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
algorithms = ['dqn', 'dqn2']
env_names = ['Cartpole', 'Assault', 'Breakout']

for env_name in env_names:
    
    plt.figure()
    plt.title(env_name)
    for algorithm in algorithms:
        try:
            f = open(f'log/{algorithm}/{env_name}_log.txt')
        except:
            continue
        
        lines = f.read().split('\n')[:-1]
        steps = np.array([float(line.split('\t')[0].split(' ')[1]) for line in lines]) / 1e7
        score = np.array([float(line.split('\t')[3].split(' ')[2]) for line in lines])
        rewards = np.array([float(line.split('\t')[4].split(' ')[2]) for line in lines])
        diff = np.abs(rewards - score)
        
        f.close()

        plt.plot(steps, rewards, label=algorithm)
        # plt.fill_between(steps, rewards + diff, rewards - diff, alpha=0.5)
    # plt.ylim(0, max(rewards) * 1.5)
    plt.legend()
    
plt.show()
```
