```python
import os
import matplotlib.pyplot as plt
```

```python
algo_names = ['A3C', 'PGQL', 'Q']
env_names = ['Assault', 'Breakout']
for env_name in env_names:
    plt.figure()
    plt.title(env_name)
    plt.grid()
    plt.xlabel(' '*40 + 'agent steps' + ' '*40 + '1e7')
    plt.ylabel('rewards')
    for algo_name in algo_names:
        try:
            f = open(f'./log/log_{algo_name}_{env_name}.txt')
        except:
            continue
            
        lines = f.read().split('\n')
        
        steps = list()
        rewards = list()
        for line in lines:
            seg = line.split(' ')
            if len(seg) != 11: continue
            steps.append(float(seg[-4][:-2]) / 10)
            rewards.append(float(seg[-1]))
        
        plt.plot(steps, rewards, label=algo_name)
    plt.legend()
plt.show()
# os.listdir('./log')
```
